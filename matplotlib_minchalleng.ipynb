```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_excel('Superstore.xlsx')
df
```

```python
flower_df = pd.read_json('iris.json')
```

```python
flower_df
```

```python
pivot_df = df.pivot_table('Sales', index='Category', aggfunc='sum')      
pivot_df
```

```python
pivot_df.plot( kind = 'bar')
plt.ylabel('Sales')
plt.title('Total Sales by Category')
plt.legend()
plt.show()
```

```python
pivot_df = df.pivot_table('Sales', index='Sub-Category', aggfunc='sum')      
pivot_df
```

```python
pivot_df.plot( kind = 'barh')
plt.ylabel('Subcategory')
plt.title('Sales by Subcategory')
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']
plt.legend('$')
plt.show()
```

```python
pivot_df = df.pivot_table('Sales', index='Region', aggfunc='sum')      
pivot_df
```

```python
pivot_df.plot.pie(y='Sales', figsize=(4,4), legend=0,  autopct='%.1f%%')
plt.ylabel('')
```

```python
pivot_df.plot.pie(y='Sales', figsize=(4,4), legend=0)

centre_circle = plt.Circle((0, 0), 0.9, fc='white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)

#save file before show
plt.savefig("squares.png")

# Displaying Chart
plt.show()
```
